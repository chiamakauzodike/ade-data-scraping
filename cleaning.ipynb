```python
import pandas as pd
```

```python
df = pd.read_csv("universities.csv")
```

```python
#extract data integers only
df["Founded"] = df["Founded"].str.extract("(\d{4})")
```

```python
df['Founded'].values
```

```python
df['Founded'] = df['Founded'].astype(int)
```

```python
df["Enrollment"].values
```

```python
df["Enrollment"] = df["Enrollment"].str.extract("(\d{1,3}(?:,\d{1,3})*)")
```

```python
df["Enrollment"].values
```

```python
df["Enrollment"] = df["Enrollment"].str.replace(',','')
```

```python
df
```

```python
df["Enrollment"] = df["Enrollment"].astype(int)
```

```python
df.info()
```

```python
df
```

```python
# Split Location into City and Country
df[["City", "Country"]] = df["Location"].str.split(",", n=1, expand=True)

# Clean and fill empty countries with city value
df["Country"] = df["Country"].fillna(df["City"]).str.strip()
df["City"] = df["City"].str.strip()

# Drop original Location column
df = df.drop(columns=["Location"])

# Reorder columns: insert City and Country after second column (index 2)
cols = df.columns.tolist()
# Remove City and Country so we can re-insert at desired location
cols.remove("City")
cols.remove("Country")
# Insert City and Country at index 2 and 3
cols[2:2] = ["City", "Country"]
df = df[cols]
```

```python
df.to_csv("universities-clean.csv", index=False)
```
