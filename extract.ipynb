```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://en.wikipedia.org/wiki/List_of_largest_universities_and_university_networks_by_enrollment")
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content)
```

```python
table = soup.find("table", attrs={"class": "sortable"})
```

```python
trs = table.findAll("tr")
```

```python
trs[0]
```

```python
columns = []
for items in trs[0].findAll("th"):
    print(items.text.strip())
    columns.append(items.text.strip())
```

```python
columns[-1] = "Link"
```

```python
columns = list(map(lambda x: x.text.strip(), trs[0].findAll("th")))
```

```python
columns[-1] = "Link"  #renmes Ref to Link on the last column
```

```python
columns
```

```python
rows = trs[1:]
```

```python
rows[0]
```

```python
rows[0].findAll("td")
```

```python
list(map(lambda x: x.text.strip(), rows[0].findAll("td")))
```

```python
rows[0].findAll("td")[1].a.attrs["href"]
```

```python
def extract_row(tr):
    row_soup_list = tr.findAll("td")
    row = list(map(lambda x: x.text.strip(), row_soup_list))
    link = row_soup_list[1].a.attrs["href"].lstrip("/")
    print("Link: ", link)
    row[-1] = f"https://en.wikipedia.org/{link}" if link else ""
    return row
```

```python
extract_row(rows[0])
```

```python
data = list(map(extract_row, rows))
```

```python
data
```

```python
#save as json or csv
import csv
import json
import pandas as pd
```

```python
with open("universities.csv", "w") as f:
    csvwriter = csv.writer(f)
    csvwriter.writerow(columns)
    csvwriter.writerows(data)
          
```

```python
with open('universities.json', 'w') as f:
    json.dump(data, f, indent=4)
```
